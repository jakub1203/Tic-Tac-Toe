```python
import numpy as np
import graphics
from graphics import *
from numpy import *
```

```python
def winning(board):
    global winner, end
    c = [2, 2, 2]
    c1 = [2, 2, 2]
    for i in range(0, 3):
        for j in range(0, 3):
            c[j] = board[i][j]
            c1[j] = board[j][i]
        if c[0] == c[1] and c[0] == c[2] and c[0] == 0:
            winner = 0
            break
        elif c1[0] == c1[1] and c1[0] == c1[2] and c1[0] == 0:
            winner = 0
            break
        elif board[0][0] == board[1][1] and board[1][1] == board[2][2] and board[2][2] == 0:
            winner = 0
            break
        elif board[0][2] == board[1][1] and board[1][1] == board[2][0] and board[2][0] == 0:
            winner = 0
            break
            

        elif c[0] == c[1] and c[0] == c[2] and c[0] == 1:
            winner = 1
            break
        elif c1[0] == c1[1] and c1[0] == c1[2] and c1[0] == 1:
            winner = 1
            break
        elif board[0][0] == board[1][1] and board[1][1] == board[2][2] and board[2][2] == 1:
            winner = 1
            break
        elif board[0][2] == board[1][1] and board[1][1] == board[2][0] and board[2][0] == 1:
            winner = 1
            break
```

```python
def x_line(x2, y2):
    global line1, line2
    line1 = Line(Point(x2 - 1.5, y2 + 1.5), Point(x2 + 1.5, y2 - 1.5))
    line2 = Line(Point(x2 - 1.5, y2 - 1.5), Point(x2 + 1.5, y2 + 1.5))

def cir(x1, y1):
    global circle
    circle = Circle(Point(x1, y1), 2)
    circle.setOutline("black")
```

```python
def boar(x1, y1, move):
    global board, x, y, end1
    for i in range(0, 3):
        if x1 > i * 5 and x1 < 5 * (i + 1):
            x = 5 * i + 2.5
            poziomo = i
        if y1 > i * 5 and y1 < 5 * (i + 1):
            y = 5 * i + 2.5
            pionowo = i
    if board[2 - pionowo][poziomo] == 2:
        board[2 - pionowo][poziomo] = move
        end1 = 0
```

```python
win = GraphWin('Tic Tac toe', 500, 500)
win.setBackground('lightgreen')
win.setCoords(0, 0, 15, 15)
for i in range(1, 3):
    ly = Line(Point(0, 5 * i), Point(15, 5 * i))
    ly.draw(win)
    lx = Line(Point(5 * i, 0), Point(5 * i, 15))
    lx.draw(win)

end = 1
end1 = 1
winner = 2
full = 1
board = np.array([[2, 2, 2],
                  [2, 2, 2],
                  [2, 2, 2]])

while end:
    #FIRST PLAYER
    
    end1 = 1
    while end1:
        position1 = win.getMouse()
        x1 = position1.getX()
        y1 = position1.getY()
        boar(x1, y1, 0)
        
    cir(x, y)
    circle.draw(win)
    winning(board)
    if winner == 0:
        break
    
    full +=1
    if full == 6:
        break
        
    #SECOND PLAYER   
    end1 = 1
    while end1:
        position2 = win.getMouse()
        x2 = position2.getX()
        y2 = position2.getY()
        boar(x2, y2, 1)
        
    x_line(x, y)
    line1.draw(win)
    line2.draw(win)
    winning(board)
    if winner == 1:
        break
    
if winner == 0:
    tekst1 = Text(Point(7.5, 7.5), 'X WON, press q to quit')
    tekst1.draw(win)
elif winner == 1:
    tekst1 = Text(Point(7.5, 7.5), 'X WON, press q to quit')
    tekst1.draw(win)
else:
    tekst1 = Text(Point(7.5, 7.5), 'Draw, press q to quit')
    tekst1.draw(win)
if win.getKey() == "q":
    win.close()
```
